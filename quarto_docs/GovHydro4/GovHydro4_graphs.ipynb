```python
#GovHydro4FrancisPeltonTurbine

import matplotlib.pyplot as plt 

francis_Gv = [0.1, 0.4, 0.5, 0.7, 0.8, 0.9, 1]
francis_Pgv = [0, 0.42, 0.56, 0.8, 0.9, 0.97, 1]

kaplan_Gv = [0.1, 0.4, 0.5, 0.7, 0.8, 0.9, 1]
kaplan_Pgv = [0, 0.35, 0.468, 0.796, 0.917, 0.99, 1]

simple_Gv  = [0, 1]
simple_Pgv = [0, 1]

fig, ax1 = plt.subplots()
ax1.plot(francis_Gv, francis_Pgv, marker='x', color='green', markeredgecolor='black',label='Francis,Pelton')
ax1.plot(kaplan_Gv, kaplan_Pgv, marker='x', color='skyblue', markeredgecolor='black', label='Kaplan')
ax1.plot(simple_Gv, simple_Pgv, marker='x', color='orange', markeredgecolor='black', label='simple')
plt.xlabel('Gate Opening ($G_\\mathrm{v}$) in Pu')
plt.ylabel('Turbine Power ($P_\\mathrm{gv}$) in Pu')
plt.grid()
plt.legend()

plt.show()
fig.savefig('./drawings/fig-gate-head-flow-characteristics-output-1.png')
```

```python
# GovHydro4KaplanTurbine

kaplan_Gv = [0.1, 0.4, 0.5, 0.7, 0.8, 0.9, 1]
kaplan_Pgv = [0, 0.35, 0.47, 0.796, 0.917, 0.99, 1]
kaplan_Bgv = [0, 0, 0.1, 0.667, 0.8, 0.9, 1]

fig, ax1 = plt.subplots()
ax1.plot(kaplan_Gv, kaplan_Pgv, marker='x', color='green', markeredgecolor='black', label='Turbine Power') # Plotting the turbine power 
ax1.set_xlabel('Gate Opening ($G_\\mathrm{v}$) in Pu')
ax1.set_ylabel('Turbine Power ($P_\\mathrm{gv}$) in Pu')
ax1.tick_params(axis='y')
ax2 = ax1.twinx() # Creating a the blade stroke
ax2.plot(kaplan_Gv, kaplan_Bgv, marker='x', color='skyblue', markeredgecolor='black',label='Blade Stroke')
ax2.set_ylabel('Gate Blade Servo Stroke ($B_\\mathrm{gv}$) in Pu')
ax2.tick_params(axis='y')
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2)
plt.show()
```
